```python
# import all required libraries
import pandas as pd # for data manipulaton 
import numpy as np # for mathematical calculations
import matplotlib.pyplot as plt # for data visualization 
import seaborn as sns # for advanced data visualization
from sklearn.model_selection import train_test_split # for split the data into train and testing
from sklearn.linear_model import LinearRegression # Fitting Multiple Linear Regression to the Training set
from sklearn import metrics # for Evalute the model

import warnings
warnings.filterwarnings("ignore")
```

```python
# load the data set
wal_feature = pd.read_csv(r"/content/features.csv") # feature data set
wal_feature
```

```python
wal_train = pd.read_csv(r"/content/train.csv") # train data set
wal_train
```

```python
wal_store = pd.read_csv(r'/content/stores.csv') # stores data set
wal_store
```

# Merging the train dataset, feature dataset and stores data set based on "store" 

```python
wallmart = wal_train.merge(wal_feature, how = 'left').merge(wal_store, how ='left')
wallmart
```

```python
#wallmart = wallmart.sample(50000)
#wallmart
```

# data preprocessing

```python
wallmart.info() # summary of the data
```

```python
# converting Date column from object to data time 
wallmart.Date.dtype
 
```

```python
date = pd.to_datetime(wallmart['Date'])
```

```python
date.dt.month[:] # only shows the month of the data
```

```python
date.dt.year # only shows year of the date
```

```python
date.dt.day # only shows day of the date
```

```python
# only consider month
wallmart['Month'] = date.dt.month
wallmart.head()
```

```python
# Drop Date column from wallmart data set
wallmart.drop('Date', axis = 1, inplace = True)
```

```python
wallmart.IsHoliday.dtype  # check the data type
```

```python
def convert (x): 
    if x == True:
        return 1
    else:
        return 0
```

```python
# convert IsHoliday into 0 and 1
wallmart['IsHoliday'] = wallmart["IsHoliday"].apply(convert)
wallmart['IsHoliday']
```

```python
# convert Type column in to dummies
wallmart['Type'].value_counts()
```

```python
type = pd.get_dummies(wallmart['Type'], drop_first = True) # convert Type column into dummy variables
type.head()
```

```python
# concatnate the dummy variables into wallmart data set
wallmart = pd.concat([wallmart, type], axis = 1) 
wallmart.head()
```

```python
# Drop the type column
wallmart.drop('Type', axis = 1, inplace = True)
wallmart.head()
```

```python
wallmart.isna().sum() # check the null values
```

```python
## replace the null values with '0'
wallmart.fillna(value = 0, inplace = True)
wallmart.isna().sum()
```

```python
 wallmart.duplicated().sum() # check duplicates in the data
```

# Univariate analysis

```python
wallmart['Store'].value_counts(ascending = True)
```

```python
## distribution plot for each column
for column in wallmart:
    sns.distplot(wallmart[column])
    plt.title(column)
    plt.show()
```

```python
## check correlation between each column
cor = wallmart.corr()
```

```python
## heat map of correlation matrix
plt.figure(figsize =(15, 10))
sns.heatmap(cor,annot = True)
plt.show()
```

```python
## create feature matrixs
independent = wallmart.drop('Weekly_Sales', axis = 1) # input variables
independent
```

```python
dependent = wallmart['Weekly_Sales'] # output variable
dependent
```

```python
# spliting the data into train and test
x_train, x_test, y_train, y_test = train_test_split(independent, dependent, test_size = 0.2, random_state = 5)
x_train.shape
x_test.shape
```

```python
## build the linear regression
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
mb = regressor.fit(x_train, y_train)
 
```

```python
# Predicting the Test set results
y_pred = mb.predict(x_test) 
```

```python
## residual values
test_res = y_test - y_pred 
test_res
```

```python
from sklearn import metrics ## model evaluation
mse = metrics.mean_squared_error(y_test, y_pred)
mse
```

```python
np.sqrt(mse) # RMSE value
```
